# Comparing counting elements in a list fastest method is using Counter from the collections module

```python
import random 
import string 
```

```python
?random.choice
```

```python
arr = random.choices(string.ascii_lowercase, k=10000)
print(arr, end="")
```

```python
dic = {}
for i in arr:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1
print(dic)
```

```python
%%timeit 
dic = {}
for i in arr:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1
```

```python
# Another way to do the above is to use the collections library 
# https://www.geeksforgeeks.org/python-collections-module/
from collections import defaultdict, Counter
```

```python
?defaultdict
```

```python
dic_default = defaultdict(int)
for i in arr:
    dic_default[i] += 1
print(dic_default)
```

```python
%%timeit
dic_default = defaultdict(int)
for i in arr:
    dic_default[i] += 1
```

```python
Counter(arr)
```

```python
%%timeit 
Counter(arr)
```
